# Generate dummie data

```python
import pandas as pd
import numpy as np
import random
from pandas import get_dummies


hostnames=[]
ips=[]
agencys=[]
location=[]
oss=[]
aid=[]
alist=['aaa','bbb','ccc']
olist=['win','server','centos']
llist=['e','a','k']


def pic():
    return random.randint(0,2)
def make_ip():
    myip = ""+str(random.randint(99,254))+"."+str(random.randint(99,254))+"."+str(random.randint(99,254))+"."+str(random.randint(99,254))
    return myip
def make_host():
    myhost="Server"+str(random.randint(0,1999))+"Bob"
    return myhost
for i in range(2500,2625):
    aid.append(i)
    
for i in range(0,250):
    ips.append(make_ip())
    hostnames.append(make_host())
    agencys.append(alist[pic()])
    location.append(llist[pic()])
    oss.append(olist[pic()])
cmdb = pd.DataFrame({'name':hostnames[0:200],'ip':ips[0:200],'os':oss[0:200],'location':location[0:200],'agency':agencys[0:200]})
log = pd.DataFrame({'name':hostnames[150:250],'ip':ips[150:250],'os':"",'location':"",'agency':"",'id':aid[0:100]})

    

```
